### Time Series Simulation

```python
# Adapter from Pandas Time_Range Docs
rng = pd.date_range('1/1/2011', periods = 72, freq = 'H')
```

```python
rng
```

```python
ts = pd.DataFrame(np.random.poisson(15, len(rng)), index = rng, columns=['admissions'])
```
